```python
import math
import matplotlib.pyplot as plt

#Declaring constants
VOLUME_CHESTNUT_BUN = 0.0005 #This is made up from an assumption
SOLAR_SYSTEM_RADIUS = 4.5e12 #This is an approximation of solar system's radius in meters (Pluto's orbit)

VOLUME_SOLAR_SYSTEM = (4/3) * math.pi * (SOLAR_SYSTEM_RADIUS ** 3)

def minutes_to_fill_solar_system(initial_volume, system_volume):
  """
  initial_volume- initial volume in cubic meters
  system_volume- system's volume in cubic meters
  elapsed_minutes - minutes needed to exceed the system volume
  volumes - list recording volume at each step
  """
  volume = initial_volume
  elapsed_minutes = 0
  volumes = [volume]
  while volume < system_volume:
    elapsed_minutes += 5
    volume *= 2
    volumes.append(volume)

  return elapsed_minutes, volumes

#Running the simulation
elapsed_minutes, volumes = minutes_to_fill_solar_system(VOLUME_CHESTNUT_BUN, VOLUME_SOLAR_SYSTEM)

#System output
print("Minutes it takes cover the solar system: {} minutes".format(elapsed_minutes))
#Equivalence to hours and days
print("Which is {:.2f} hours, or {:.2f} days".format(elapsed_minutes / 60, elapsed_minutes / 1440))

#Visualizing using graphs
plt.figure(figsize=(10,6))
plt.plot(range(0, elapsed_minutes + 5, 5), volumes)
plt.title("Growth of Chestnut Buns (Baibain Effect)")
plt.xlabel("Minutes")
plt.ylabel("Total Volume (cubic meters)")
plt.yscale('log')
plt.grid(True)
plt.show()
```
