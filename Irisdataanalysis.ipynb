```python
#Problem 1: Data acquisition
from sklearn.datasets import load_iris
import pandas as pd

# Loading iris dataset
iris = load_iris()

# Explanatory variables (features)
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)

print("Features (X):")
X.head()

# Objective variable (label)
y = pd.DataFrame(data=iris.target, columns=["Species"])

# Displaying X.head()
X.head()
```

```python
# Displaying y.head()
y.head()
```

```python
#Problem 2: Combining data
# Combining features and target into one DataFrame
df = pd.concat([X, y], axis=1)

# Displaying first few rows
df.head()
```

```python
#Problem 3: Checking the data

# 1. Displaying the 4th sample (index 3)
print("4th sample:\n", df.iloc[3])

# 2. Count of each label
print("\nSample count by label:\n", df['Species'].value_counts())

# 3. Checking for missing values
print("\nMissing values:\n", df.isnull().sum())

# 4. Displaying statistical summary
print("\nDescriptive statistics:\n", df.describe())
```

#Problem 4: Examining the iris dataset itself
- Collected by Edgar Anderson and popularized by R.A. Fisher.
- 150 iris flower samples from three species: Setosa (0), Versicolor (1), Virginica (2).
- Collected in the 1930s from the Gasp Peninsula, Canada.
- It is a classic dataset for testing classification algorithms.


```python
#Problem 5: Extracting the required data
# 1. Extracting 'sepal width' column in two ways
print(df['sepal width (cm)'].head())        # Method 1
print(df.iloc[:, 1].head())                 # Method 2

# 2. Extracting 50th to 99th data
print(df.iloc[50:100])

# 3. Extracting petal_length of rows 50 to 99
print(df.loc[50:99, 'petal length (cm)'])

# 4. Extracting rows with petal_width == 0.2
print(df[df['petal width (cm)'] == 0.2])

# Explanation
# .loc[] accesses rows/columns by labels
# .iloc[] accesses rows/columns by integer positions
```

```python
#Problem 6: Creating a diagram
import matplotlib.pyplot as plt
import seaborn as sns

# Pie chart
label_counts = df['Species'].value_counts()
labels = iris.target_names
plt.figure(figsize=(6,6))
plt.pie(label_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Iris Species Distribution")
plt.show()

# Box plot for each feature by label
for col in iris.feature_names:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Species', y=col, data=df)
    plt.title(f"Boxplot of {col} by Species")
    plt.show()

# Violin plot for each feature by label
for col in iris.feature_names:
    plt.figure(figsize=(6,4))
    sns.violinplot(x='Species', y=col, data=df)
    plt.title(f"Violin plot of {col} by Species")
    plt.show()

# How they differ and advantages each has:
# Boxplot shows median, quartiles, and outliers.
# Violin plot shows KDE (distribution) plus boxplot in one plot.
#Box Plots are better for simplicity and summary stats (e.g., group comparisons).
# Violin plots are better at showing distribution shape.
```

```python
#Problem 7: Confirming the relationship between features
# Scatter plots (all pairs)
import itertools

pairs = list(itertools.combinations(iris.feature_names, 2))
for x_col, y_col in pairs:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=x_col, y=y_col, hue='Species', data=df, palette='Set2')
    plt.title(f"{x_col} vs {y_col}")
    plt.show()

# Scatterplot matrix
sns.pairplot(df, hue='Species', diag_kind='hist', palette='husl')
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()

# Correlation matrix
corr_matrix = df[iris.feature_names].corr()
print("Correlation Matrix:\n", corr_matrix)

# Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap of Correlation Coefficients")
plt.show()
```

Problem 8: Explaining the results of visualized graphs and tables
#Discussion of results from the data:

**Label Distribution:**
*   There are 50 samples (33.3%) of each iris species.




**Feature Descriptions**:
* The values of sepals length and petal length are more than their widths.
* Setosa is typically a lot smaller in terms of the size of its petals in comparison to the others.

**Box/Violin Plots**:
* Setosa is plainly different in petal length and width (little, narrow range).
* Versicolor and Virginica overlap more, nevertheless, differ in medians and distributions.

**Scatter Plots & Pairplot**:
* All 3 species are well separated by petal length vs petal width.
* The sepal measurements are not useful in classification.
* Setosa can be linearly separated to the rest.

**Correlation Matrix**:
* Petal length and petal width are very strongly and positively related (~0.96).
* The width of the sepal is weakly associated with the width/length of the petal.

**Conclusion**:
* The length and width of the petal are the best measuring tools to differentiate between species.
* Data visualization can help uncover meaningful structure in the data, as well as guide the choice of features to use in machine learning.
