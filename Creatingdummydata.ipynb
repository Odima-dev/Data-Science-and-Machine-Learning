```python
#1. Creating random numbers
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mean1 = [-3, 0]
covariance = [[1.0, 0.8],
              [0.8, 1.0]]

#Creating 500 random numbers
data1 = np.random.multivariate_normal(mean1, covariance, 500)
print("Created random Data1:", data1.shape)
print("\n")

#2. Visualization using scatter plots
plt.figure(figsize=(8, 6))
plt.scatter(data1[:, 0], data1[:, 1])
plt.title('Scatter Plot of Data1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
print("\n")

#3. Visualization using histograms
# Histogram of x1
plt.figure(figsize=(8, 4))
plt.hist(data1[:, 0], bins=30, color='blue', alpha=0.7)
plt.title('Histogram of Data1 (X1)')
plt.xlabel('x1')
plt.ylabel('Frequency')
plt.xlim(-7, 3)
plt.grid(True)
plt.show()

# Histogram of x2
plt.figure(figsize=(8, 4))
plt.hist(data1[:, 1], bins=30, color='green', alpha=0.7)
plt.title('Histogram of Data1 (X2)')
plt.xlabel('x2')
plt.ylabel('Frequency')
plt.xlim(-7, 3)
plt.grid(True)
plt.show()
print("\n")

#4. Adding data
mean2 = [0, -3]
data2 = np.random.multivariate_normal(mean2, covariance, 500)
print("Created random Data2:", data2.shape)

# Visualizing the data for questions 1 and 4 as one scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data1[:, 0], data1[:, 1], label='Class 0', alpha=0.7)
plt.scatter(data2[:, 0], data2[:, 1], label='Class 1', alpha=0.7)
plt.title('Scatter Plot of Data1 and Data2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()

#5. Data Combining
combined_data = np.vstack((data1, data2))
print("Shape of combined_data:", combined_data.shape)

#6. Labeling using 0 for data1, 1 for data2
labels1 = np.zeros((500, 1))
labels2 = np.ones((500, 1))
#Combining vertically
combined_labels = np.vstack((labels1, labels2))
# Concatenating data and labels horizontally
final_dataset = np.hstack((combined_data, combined_labels))
print("Shape of final_dataset:", final_dataset.shape)

print(final_dataset[:6])
```
