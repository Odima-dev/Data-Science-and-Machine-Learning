```python
#1. Creating the function to calculate grains
def compute_sorori_shinzaemon(n_days=100):
  list_n_grains = []
  list_total_grains = []
  grains_today = 1
  total_grains = 0
  for _ in range(n_days):
    list_n_grains.append(grains_today)
    total_grains += grains_today
    list_total_grains.append(total_grains)
    grains_today *= 2


  return list_n_grains, list_total_grains

list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=10)
#Displaying output for n_days= 10
print(compute_sorori_shinzaemon(n_days=10))

#2. Visualizing through graph
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
#Grains per day
plt.plot(list_n_grains, label='Grains per Day', color='red', linestyle='dotted', linewidth=2)
#Cumulative grains
plt.plot(list_total_grains, label='Cumulative Grains', color='blue', linestyle='solid', linewidth=2)

plt.title("Sorori Shinzaemon's Rice Story")
plt.xlabel("Day")
plt.ylabel("Number of Rice Grains")
plt.legend()

plt.tick_params(labelsize=12)

plt.grid(True)
plt.show()

#3. Estimating how many days you can live on a grain of rice
# A typical person consumes around 18,000-19,000 grains of rice per day
def estimate_lifetime(total_grains, grains_per_day=18000, num_people=1):
  """
  total_grains - total number of available grains (default set to 18000)
  grains_per_day - amount of grains one can consume per day
  num_people - number of people sharing the rice (default set to 1)
  """
  grains_needed_per_day = grains_per_day * num_people
  days = total_grains // grains_needed_per_day
  return days

# Example of a scenario
list_n_grains, list_total_grains = compute_sorori_shinzaemon(100)
final_total = list_total_grains[-1]  # Total after 100 days

days_survived = estimate_lifetime(final_total, grains_per_day=18000, num_people=3)
print(f"3 people can survive for {days_survived} days with the rice received.")



```
